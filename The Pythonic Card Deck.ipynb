```python
import collections

Card = collections.namedtuple('Card', ['rank', 'suit'])

class FrenchDeck:
    ranks = [str(n) for n in range(2, 11)] + list('JQKA')
    suits = 'spades diamonds clubs hearts'.split()
    
    def __init__(self):
        self.cards = []
        for suit in self.suits:
            for rank in self.ranks:
                self.cards.append(Card(rank, suit))

    def __len__(self):
        return len(self.cards)

    def __getitem__(self, position):
        return self.cards[position]

deck = FrenchDeck()
print(len(deck))  # Output: 52
print(deck[0])  # Output: Card(rank='2', suit='spades')
print(deck[-1])  # Output: Card(rank='A', suit='hearts')
print(deck[12:15])  # Output: [Card(rank='3', suit='spades'), ...]
```

```python
from random import choice
print(choice(deck))  # Output: Random card from the deck
print(deck[:4])  # Output: First four cards in the deck


for card in deck:
    print(card)

for card in reversed(deck):
    print(card) 

Card(rank='K', suit='spades ') in deck  # Output: False?
```

```python
# WITHOUT magic methods - you have to access .cards directly

class SimpleDeck:
    ranks = [str(n) for n in range(2, 11)] + list('JQKA')
    suits = 'spades diamonds clubs hearts'.split()
    
    def __init__(self):
        self.cards = []
        for suit in self.suits:
            for rank in self.ranks:
                self.cards.append(Card(rank, suit))

# Without __len__, you must call len() on .cards
simple_deck = SimpleDeck()
print(len(simple_deck.cards))  # len(simple_deck) won't work!

# Without __getitem__, you must use .cards[index]
print(simple_deck.cards[0])  # deck[0] won't work!
print(simple_deck.cards[-1])  # deck[-1] won't work!
print(simple_deck.cards[12:15])  # deck[12:15] won't work!

# Without __getitem__, you must iterate over .cards
for card in simple_deck.cards:  # for card in simple_deck won't work!
    print(card)

# Without __getitem__, reversed() won't work on the deck
for card in reversed(simple_deck.cards):  # reversed(simple_deck) won't work!
    print(card)
```
